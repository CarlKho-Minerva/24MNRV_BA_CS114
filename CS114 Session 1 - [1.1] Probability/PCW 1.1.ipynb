# Set notation cheat sheet

| Description | Notation | Visualization |
|-------------|----------|---------------|
| Sample space. The set of all possible outcomes. | S (In some other textbooks, it is written using the Greek symbol .) | |
| A set. A collection of outcomes. | A, B, or any other capital letter. | |
| Complement of A. The set of all outcomes that are not in A. | Ac (Written as A^c in TeX.) | |
| Empty set. The set with no outcomes. | ∅ or {} (Written as \varnothing in TeX.) | |
| Subset. A is a subset of B if all the outcomes in A are also in B. | A ⊆ B (Written as \subseteq in TeX.) | |
| Union. The set that contains all the outcomes that are in at least one of the given sets. | Two sets: A ∪ B (Written as \cup in TeX.)<br>Three sets: A ∪ B ∪ C<br>Many sets: ⋃<sub>i=1</sub><sup>n</sup> A<sub>i</sub> (Written as \bigcup in TeX.) | |
| Intersection. The set that contains all the outcomes that are in all of the given sets. | Two sets: A ∩ B (Written as \cap in TeX.)<br>Three sets: A ∩ B ∩ C<br>Many sets: ⋂<sub>i=1</sub><sup>n</sup> A<sub>i</sub> (Written as \bigcap in TeX.) | |
| Difference. The set that contains all outcomes in one set that are not in the other set. | The difference of B and A: B \ A. This is not symmetric so B \ A ≠ A \ B. | |

The source for producing the visualizations is in this Python notebook.

```python
# Sample Space ("S")

plt.figure(figsize=(4, 4))
plt.xlim(-0.666, 0.666)
plt.ylim(-0.666, 0.666)
plt.axis("on")
plt.xticks([])
plt.yticks([])
plt.text(
    2 / 3 - 2 / 30,
    2 / 3 - 2 / 30,
    "S",
    horizontalalignment="right",
    verticalalignment="top",
)
plt.gca().set_facecolor("cornflowerblue")
plt.show()
```

```python
# A Set ("A")

plt.figure(figsize=(4, 4))
set_sizes = {"10": 0.4, "01": 0, "11": 0}
v = venn.venn2(
    subsets=set_sizes,
    set_labels=("", ""),
    alpha=1,
    subset_label_formatter=lambda x: {0.4: "A"}.get(x, ""),
)
v.get_patch_by_id("10").set_color("cornflowerblue")
c = venn.venn2_circles(subsets=set_sizes, linewidth=1)
plt.xlim(-2 / 3, 2 / 3)
plt.ylim(-2 / 3, 2 / 3)
plt.axis("on")
plt.text(
    2 / 3 - 2 / 30,
    2 / 3 - 2 / 30,
    "S",
    horizontalalignment="right",
    verticalalignment="top",
)
plt.show()
```

```python
# Complement of A ("A^c")

plt.figure(figsize=(4, 4))
set_sizes = {"10": 0.4, "01": 0, "11": 0}
v = venn.venn2(
    subsets=set_sizes,
    set_labels=("", ""),
    alpha=1,
    subset_label_formatter=lambda x: {0.4: "A"}.get(x, ""),
)
v.get_patch_by_id("10").set_color("white")
c = venn.venn2_circles(subsets=set_sizes, linewidth=1)
plt.xlim(-2 / 3, 2 / 3)
plt.ylim(-2 / 3, 2 / 3)
plt.axis("on")
plt.text(
    2 / 3 - 2 / 30,
    2 / 3 - 2 / 30,
    "S",
    horizontalalignment="right",
    verticalalignment="top",
)
plt.gca().set_facecolor("cornflowerblue")
plt.show()
```

```python
# Subset A of B ("A \subset B")

plt.figure(figsize=(4, 4))
set_sizes = {"10": 0, "01": 0.6, "11": 0.4}
v = venn.venn2(
    subsets=set_sizes,
    set_labels=("", ""),
    alpha=1,
    subset_label_formatter=lambda x: {0.4: "A", 0.6: "B"}.get(x, ""),
)
v.get_patch_by_id("11").set_color("royalblue")
v.get_patch_by_id("01").set_color("cornflowerblue")
c = venn.venn2_circles(subsets=set_sizes, linewidth=1)
plt.xlim(-2 / 3, 2 / 3)
plt.ylim(-2 / 3, 2 / 3)
plt.axis("on")
plt.text(
    2 / 3 - 2 / 30,
    2 / 3 - 2 / 30,
    "S",
    horizontalalignment="right",
    verticalalignment="top",
)
plt.gca().set_facecolor("white")
plt.show()
```

```python
import random


# The simulation function
def drop_one_ball():
    position = 0
    for step in range(10):
        position += random.choice([0, 1])
    return position
```

```python
# TASK: Run the simulation 1000 times and record the results.
# Hint: Write a for-loop and use results.append().
results = []
for step in range(100):
    results.append(drop_one_ball())
```

```python
# Here we plot the results. You don't have to modify this code.
# Hint: This won't work unless you completed the code above.
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8, 6))
plt.hist(results, bins=np.arange(11) - 0.5, edgecolor="white")
plt.xticks(range(11))
plt.xlim(-0.6, 10.6)
plt.title("Histogram of simulation results")
plt.xlabel("Final position")
plt.ylabel("Count")
plt.show()
```
