```python
'''
Your explanation here.
'''

import numpy as np
import matplotlib.pyplot as plt

# Use comb to find the number of combinations, comb(n, k) means "n choose k".
from scipy.special import comb

N = 500
n = 30
m = 500

# Count the number of times you observe k elk that have already been tagged.
# All values from k = 0 to k = m are possible.
counts = np.zeros(m + 1, dtype=int)
trials = 10000
for _ in range(trials):

    # Write your simulation here
    first_sample = np.random.choice(N,size= n, replace = False) # tagged elks
    second_sample = np.random.choice(N, size = m, replace = False) # recaptured elks
    k = len(set(second_sample).intersection(set(first_sample))) #size of intersection
    counts[k] += 1

plt.figure(figsize=(8, 6))
k = np.arange(m + 1)
plt.bar(k, counts / trials, label='simulated')

# Also plot your answer from Question 1 here.
# Do the theoretical and simulated results match?

plt.plot(k, comb(n, k) * comb(N-n, m - k) / comb(N, m), linestyle = 'none', marker= 'o', markersize = 8, color = 'orange')
plt.legend()
plt.xlabel('count of previously tagged elk')
plt.ylabel('probability')
plt.xlim([0,n+1])

plt.show()
```
