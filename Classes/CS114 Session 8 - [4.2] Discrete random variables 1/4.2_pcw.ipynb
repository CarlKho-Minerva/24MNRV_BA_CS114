```python
import random

# This could be any value between 0 and 1, representing the probability
# of a positive outcome.
p = 0.75
test = random.uniform(0, 1)
if test < p:
    outcome = "positive"
else:
    outcome = "negative"

print(f"outcome '{test}'")
```

```python
import scipy.stats as sts

p = 0.5
distribution = sts.bernoulli(p)  # Create a Bernoulli distribution
print("p(X = 0) =", distribution.pmf(0))  # Display the values of the PMF
print("p(X = 1) =", distribution.pmf(1))
print("Random samples:", distribution.rvs(5))
```

```python
import scipy.stats as sts
import matplotlib.pyplot as plt
import numpy as np

# Create a Binomial distribution and set its parameter values
p = 0.75
n = 10
distribution = sts.binom(n, p)

# Generate random values from the distribution
print("Random samples:", distribution.rvs(5))

# Plot the PMF
k = np.arange(n + 1)
plt.plot(k, distribution.pmf(k), "ko")
plt.show()
```

```python
import scipy.stats as sts
import matplotlib.pyplot as plt
import numpy as np

# Create a Hypergeometric distribution
N = 20  # There are 20 marbles in the jar
K = 7  # 7 of the marbles are amber-colored
n = 12  # We pick 12 marbles and count how many are amber-colored
distribution = sts.hypergeom(N, K, n)

# Generate random values from the distribution
print("Random samples:", distribution.rvs(5))

# Plot the PMF
k = np.arange(n + 1)
plt.plot(k, distribution.pmf(k), "ko")
plt.show()
```
